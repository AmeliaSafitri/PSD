## **Cara Pengukuran 577 Atribut (Secara Spesifik)**

1. Proses Akuisisi Awal: Video dan 
Garis Luar (Outline)

    - Asal Data: Pengukuran diambil dari rekaman video lalu lintas.

    - Langkah Komputer: mengidentifikasi objek mobil dan kemudian mengekstrak Garis Luar (Outline) atau Kontur bentuk mobil dari frame video tersebut

    - Hasil: Pada titik ini, yang dimiliki adalah bentuk 2D (dua dimensi) dari profil mobil.

2. Menemukan Titik Pusat (Centroid Calculation)

    - Setelah profil 2D mobil diisolasi dan dinormalisasi, komputer harus menentukan Titik Pusat Geometri (Centroid) dari bentuk tersebut.
    
    - Titik ini akan berfungsi sebagai titik referensi tetap (titik 0,0) untuk semua pengukuran. Ini penting agar pengukuran tidak berubah jika mobil sedikit bergeser posisinya.

3. Membagi Lingkaran Menjadi 577 Sektor (Angular Sampling)

    - membayangkan sebuah lingkaran penuh (360 derajat) yang berpusat di Centroid yang telah ditemukan.
                        
    -  Lingkaran ini kemudian dibagi menjadi 577 sudut atau interval yang sama persis (360 derajat/577).
                                
    - Setiap pembagian sudut ini akan menghasilkan garis lurus yang memancar dari titik pusat ke garis luar mobil.

4. Pengukuran Jarak (Distance Measurement)

    - Mulai dari sudut pertama (yang akan menjadi att1), komputer melakukan pengukuran:
                                            
    - Jarak diukur dari Titik Pusat (Centroid) hingga ke Garis Luar (Kontur) mobil di sepanjang garis sudut pertama tersebut.
                                                    
    - Nilai jarak ini disimpan sebagai att1.
                                                        
    - Proses ini diulang untuk sudut kedua (yang menjadi att2), sudut ketiga (att3), dan seterusnya, hingga sudut ke-577 (att557)

5. Menghasilkan Deret 1D
                                                            
    - Komputer kini memiliki urutan 577 angka yang bermakna secara geometris.
                                                                  
    - Nilai Jarak Pendek (misalnya att(i) bernilai kecil) menunjukkan bagian mobil yang dekat dengan pusat (biasanya bagian bawah atau kap mesin rendah).
    
    - Nilai Jarak Panjang (misalnya att(j) bernilai besar) menunjukkan bagian mobil yang jauh dari pusat (biasanya bagian atap tinggi atau bumper).

**Kesimpulan**

Setiap angka dari att1 sampai att577 adalah Jarak dari pusat mobil ke garis luarnya, yang diukur secara berurutan mengelilingi seluruh profil 2D. Deret angka ini menjadi input bagi Model SVC untuk mendeteksi pola bentuk.

### Load Data

```python
import pandas as pd
import numpy as np
from scipy.io import arff
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# -------- fungsi load arff --------
def load_arff(filename):
    """
    Memuat data dari file ARFF ke dalam DataFrame, melakukan decoding byte
    ke string, dan memetakan label kelas.
    """
    data, meta = arff.loadarff(filename)
    df = pd.DataFrame(data)

    # Decode byte → string untuk kolom object
    for col in df.select_dtypes(['object']).columns:
        df[col] = df[col].str.decode('utf-8')

    df.columns = meta.names()

    # Mapping label kelas
    mapping = {
        '1': 'Sedan',
        '2': 'Pickup',
        '3': 'Minivan',
        '4': 'SUV'
    }

    label_raw = df.columns[-1]
    df["label"] = df[label_raw].astype(str).map(mapping)

    # Identifikasi kolom fitur (yang dimulai dengan 'att')
    features = [c for c in df.columns if c.startswith("att")]
    return df, features, "label"
```

**Tujuan:**

  Mengubah data mentah dari format khusus (.arff) menjadi struktur yang dapat diolah (Pandas DataFrame dan NumPy array).

**Aksi Kunci:**

 Fungsi load_arff mengurus pembacaan file ARFF, decoding nilai byte menjadi string, dan memetakan label numerik ('1', '2', dll.) ke nama kelas yang dapat dimengerti ('Sedan', 'Pickup', dll.).

### Visualisasi Data

```python
# ============= LOAD DATA =============
train_path = "/content/Car_TRAIN.arff"
test_path  = "/content/Car_TEST.arff"

df_train, feature_cols, target_col = load_arff(train_path)
df_test,  _, _ = load_arff(test_path)

print(f"Shape Data Train: {df_train.shape}")
print(f"Shape Data Test: {df_test.shape}")


# 1. Distribusi Data (TARGET CLASS)
plt.figure(figsize=(7, 5))
sns.countplot(x=target_col, data=df_train, palette="viridis",
              order=df_train[target_col].value_counts().index)
plt.title("Distribusi Kelas (Data Training)")
plt.xlabel("Tipe Mobil")
plt.ylabel("Jumlah Sampel")
plt.show()

# Distribusi Fitur (Contoh Box Plot)
# Memilih 4 fitur pertama sebagai contoh
selected_features = feature_cols[:4]
df_train_melt = pd.melt(df_train, id_vars=target_col, value_vars=selected_features,
                        var_name='Feature', value_name='Value')

plt.figure(figsize=(12, 6))
sns.boxplot(x='Feature', y='Value', hue=target_col, data=df_train_melt, palette="Set2")
plt.title("Box Plot Nilai Fitur Berdasarkan Kelas (Contoh)")
plt.xlabel("Fitur")
plt.ylabel("Nilai")
plt.legend(title='Kelas')
plt.show()
```

### Preprocessing Data


```python
# ============= PREPROCESSING =============
X_train = df_train[feature_cols].values
y_train = df_train[target_col].values

X_test = df_test[feature_cols].values
y_test = df_test[target_col].values

# Standard Scaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

print("\nPra-pemrosesan selesai. Data telah distandarisasi.")
```

**Fungsinya:**

Melakukan Standardisasi (Scaling) fitur menggunakan StandardScaler. Tujuan dari langkah ini adalah untuk memastikan semua fitur memiliki skala yang seragam, yang sangat penting untuk kinerja optimal model SVC.

### Pemodelan Data


```python
# ============= MODEL SVC =============
model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
model.fit(X_train_scaled, y_train)

print("\nPelatihan Model SVC selesai.")
```

**Fungsinya:**

Membuat sebuah objek model klasifikasi menggunakan Support Vector Machine (SVC) dari pustaka scikit-learn

### Evaluasi



```python
# ============= EVALUASI =============
y_pred = model.predict(X_test_scaled)

print("\n--- Hasil Evaluasi Kinerja ---")
print(f"Akurasi: {round(accuracy_score(y_test, y_pred) * 100, 2)} %")
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

**Fungsinya:**

Mengukur seberapa baik model memprediksi data yang belum pernah dilihat (test data).

Metode evaluasi yang digunakan:
1. Accuracy Score
  
        Presentase prediksi benar.
2. Classification Report
    
    Berisi:

        - Precision → keakuratan prediksi positif

        - Recall → kemampuan menemukan semua data positif

        - F1-score → rata-rata precision dan recall

        - Support → jumlah data per kelas

### Visualisasi Hasil (Confusion Matrix & F1-Score)

**F1-Score**

### load_data.py

Tujuan Utama: Mengubah file data mentah berformat ARFF menjadi DataFrame Pandas yang siap diproses.

- Fungsi Kunci: Memuat data dari Car_TRAIN.arff dan Car_TEST.arff menggunakan scipy.io.arff.

- Preprocessing Data: Mengkonversi data yang berbentuk byte-string menjadi string biasa.

- Output: Mengembalikan DataFrame, daftar nama fitur, dan nama kolom target.

```python
# load_data.py
import pandas as pd
from scipy.io import arff

# Nama file train & test
TRAIN_FILE_NAME = "Car_TRAIN.arff"
TEST_FILE_NAME = "Car_TEST.arff"

def load_arff_data(path):
    """
    Load file ARFF dan konversi ke DataFrame pandas.
    Mengembalikan: df, nama_label, fitur, raw
    """
    try:
        data, meta = arff.loadarff(path)
    except Exception as e:
        print(f"Error membaca ARFF {path}: {e}")
        return pd.DataFrame(), None, None, None

    df = pd.DataFrame(data)

    # Konversi byte-string -> str
    for col in df.columns:
        if df[col].dtype == object:
            df[col] = df[col].apply(lambda x: x.decode() if isinstance(x, bytes) else x)

    target = df.columns[-1]               # kolom terakhir = label
    features = df.columns[:-1].tolist()   # semua kecuali label

    return df, target, features, meta
```

### model_svc.py (Pelatihan Model & Penyimpanan)

Tujuan Utama: Melatih model klasifikasi Support Vector Classifier (SVC) dan menyimpannya agar bisa digunakan di aplikasi web.

- Pemetaan Label: Mengubah label numerik (1, 2, 3, 4) menjadi nama mobil (Sedan, SUV, dll.).

- Preprocessing Penting: Melakukan Normalisasi (Scaling) fitur menggunakan StandardScaler. Ini wajib untuk model SVC.

- Pelatihan Model: Melatih model SVC (SVC(kernel='rbf')) menggunakan data yang sudah di-skala.

- Penyimpanan (.pkl): Menyimpan model SVC, StandardScaler, dan nama-nama kolom ke dalam satu file bernama model_svc.pkl menggunakan pickle.

```python
import pandas as pd
import pickle
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from load_data import load_arff_data, TRAIN_FILE_NAME

# -------------------------------
# MAPPING LABEL
# -------------------------------
# Mengubah angka 1-4 menjadi nama mobil
label_map = {
    "1": "Sedan",
    "2": "Pickup",
    "3": "Minivan",
    "4": "SUV"
}

# -------------------------------
# LOAD DATA TRAINING
# -------------------------------
# Memuat dataset training dari file ARFF
df_train, target_col, feature_cols, _ = load_arff_data(TRAIN_FILE_NAME)

# -------------------------------
# KONVERSI LABEL KE STRING
# -------------------------------
# Ubah kolom target menjadi string sesuai label_map
df_train[target_col] = df_train[target_col].astype(str).map(label_map)

# -------------------------------
# PERSIAPAN FITUR & TARGET
# -------------------------------
# Ambil nilai fitur (X) dan target (y)
X = df_train[feature_cols].astype(float).values  # Fitur numerik
y = df_train[target_col].values                  # Target berupa string

# -------------------------------
# SCALING / PREPROCESSING FITUR
# -------------------------------
# Normalisasi fitur agar memiliki mean=0 dan std=1
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -------------------------------
# TRAIN MODEL SVC
# -------------------------------
# Membuat model Support Vector Classifier dengan kernel RBF
svc = SVC(kernel='rbf', probability=True)
svc.fit(X_scaled, y)  # Training model dengan data yang sudah diskalakan

# -------------------------------
# SIMPAN MODEL + PREPROCESSING
# -------------------------------
# Semua komponen penting disimpan dalam satu dictionary
model_dict = {
    "model": svc,           # Model SVC terlatih
    "scaler": scaler,       # Scaler untuk preprocessing fitur
    "features": feature_cols,  # Nama-nama fitur
    "target": target_col,      # Nama kolom target
    "df_test": df_train        # Test dataset (sama dengan train di sini)
}

# Simpan dictionary ke file pickle
with open("model_svc.pkl", "wb") as f:
    pickle.dump(model_dict, f)

print("✅ DONE! Model siap digunakan")
```

### app.py (Aplikasi Streamlit Interaktif)

Tujuan Utama: Membuat antarmuka web yang menampilkan evaluasi model dan memungkinkan pengguna melakukan prediksi sampel mobil.

- Pemuatan Awal: Membaca file model_svc.pkl dan mengekstrak model dan scaler yang tersimpan.

- Evaluasi Model: Menampilkan akurasi dan Classification Report model berdasarkan data uji (df_test).

- Prediksi Interaktif:

    - Memungkinkan pengguna memilih sampel data dari dropdown.

    - Mengambil fitur sampel, melakukan scaling dengan scaler yang tersimpan, dan memprediksi tipe mobil.

    - Menampilkan hasil prediksi model dan label asli sampel.

- Visualisasi: Membuat grafik deret waktu menggunakan Altair untuk memvisualisasikan nilai-nilai fitur dari sampel mobil yang dipilih.

```python
import streamlit as st
import pandas as pd
import altair as alt
import pickle
from sklearn.metrics import accuracy_score, classification_report

# -------------------------------
# LOAD MODEL
# -------------------------------
MODEL_PATH = "model_svc.pkl"

try:
    # Load model, scaler, fitur, target, dan dataset test
    model_data = pickle.load(open(MODEL_PATH, "rb"))
    model = model_data["model"]          # Model SVC
    scaler = model_data["scaler"]        # Scaler untuk normalisasi
    saved_features = model_data["features"]  # Nama-nama fitur
    saved_target = model_data["target"]      # Nama kolom target
    df_test = model_data["df_test"]      # Data test
    st.success("🟢 model_svc.pkl berhasil dimuat!")
except Exception as e:
    st.error(f"❌ Gagal load model: {e}")
    st.stop()

# -------------------------------
# APP TITLE
# -------------------------------
st.title("🚗 Prediksi Tipe Mobil")

# -------------------------------
# EVALUASI MODEL
# -------------------------------
st.header("📊 Evaluasi Model")

# Ambil fitur dan lakukan scaling
X_test_scaled = scaler.transform(df_test[saved_features])
y_test = df_test[saved_target].values

# Prediksi seluruh data test
y_pred = model.predict(X_test_scaled)

# Hitung akurasi
accuracy = (y_test == y_pred).mean() * 100
st.metric("Akurasi Data Uji", f"{accuracy:.2f}%")

# Confusion report / classification report dalam bentuk dataframe
report = pd.DataFrame(classification_report(y_test, y_pred, output_dict=True)).transpose()
st.dataframe(report)

# -------------------------------
# DATA TEST HEAD
# -------------------------------
st.header("📄 Data Uji (head)")
st.dataframe(df_test.head())  # Menampilkan 5 baris pertama dataset test

# -------------------------------
# PREDIKSI PER SAMPEL
# -------------------------------
st.header("1️⃣ Prediksi Sampel Mobil")

# Membagi layout: kolom kiri untuk select box, kolom kanan untuk hasil prediksi
col1, col2 = st.columns([1,2])

# Pilih sampel berdasarkan index
idx = col1.selectbox("Pilih ID Sampel", df_test.index.to_list())
selected = df_test.loc[idx]

# Ambil fitur sampel dan ubah ke array 2D
X_sample = selected[saved_features].astype(float).values.reshape(1, -1)

# Prediksi model untuk sampel ini
pred_label = model.predict(scaler.transform(X_sample))[0]

# Label asli dari dataset
true_label = selected[saved_target]

# Tampilkan hasil prediksi dan label asli
with col2:
    st.markdown(f"### Prediksi Model: 🚘 **{pred_label}**")
    st.markdown(f"### Label Asli: 🏷️ **{true_label}**")

# -------------------------------
# GRAFIK TIME SERIES
# -------------------------------
st.header("2️⃣ Grafik Deret Waktu Fitur Mobil")

# Input untuk memilih range fitur yang ingin ditampilkan
col3, col4 = st.columns(2)
start = col3.number_input("Mulai fitur ke-", 1, len(saved_features), 1)
end = col4.number_input("Sampai fitur ke-", 1, len(saved_features), len(saved_features))

# Jika range valid, buat chart
if start <= end:
    feats = saved_features[start-1:end]  # Ambil nama fitur sesuai range
    df_plot = pd.DataFrame({
        "Langkah Waktu": range(start, end + 1),
        "Nilai Fitur": selected[feats].astype(float).values
    })
    chart = alt.Chart(df_plot).mark_line(point=True).encode(
        x="Langkah Waktu",
        y="Nilai Fitur",
        tooltip=["Langkah Waktu", "Nilai Fitur"]
    ).properties(
        title="📈 Grafik Deret Waktu Fitur Mobil"
    ).interactive()

    st.altair_chart(chart, width='stretch')

# -------------------------------
# DETAIL DATA SAMPEL
# -------------------------------
st.header("3️⃣ Detail Data Sampel")

# Expander untuk melihat semua detail kolom sampel
with st.expander("Klik untuk melihat detail"):
    st.dataframe(pd.DataFrame(selected).T)  # Transpose agar kolom menjadi row
```
