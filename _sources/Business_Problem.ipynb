## Business Problem Dataset Iris
1. Menentukan Tujuan Bisnis
- Tujuan utama: Mengidentifikasi jenis bunga Iris (Setosa, Versicolor, Virginica) berdasarkan ukuran sepal dan petal.
- Manfaat bisnis:

 1) Petani bisa memisahkan bibit sesuai jenis.

2) Penjual bunga bisa memastikan kualitas produk sesuai pesanan.

3) Peneliti bisa menghemat waktu dan biaya dalam identifikasi spesies.
2. Membuat Rencana Proyek
- Langkah 1: Ambil data dari database iris_dataset.sql (berisi panjang & lebar sepal, petal, serta class).

- Langkah 2: Lakukan analisis deskriptif → hitung min, max, rata-rata tiap kolom untuk memahami distribusi data.

- Langkah 3: Buat visualisasi (grafik batang untuk jumlah kelas, scatter plot untuk pola sepal vs petal).

- Langkah 4: (Opsional untuk data science) Latih model klasifikasi sederhana untuk memprediksi jenis bunga berdasarkan input fitur.

- angkah 5: Implementasikan hasil analisis ke dalam dashboard (misalnya di Power BI) agar mudah digunakan oleh pihak bisnis.
3. Menilai Situasi dan Kondisi
- Kondisi dataset:

1) Jumlah data: 150 baris.

2) Jumlah kelas: seimbang (masing-masing 50 data untuk Setosa, Versicolor, Virginica).

3) Fitur: sepal_length, sepal_width, petal_length, petal_width.

- Situasi bisnis:

1) Belum ada sistem otomatis untuk identifikasi bunga.

2) Proses manual (berdasarkan pengamatan manusia) bisa salah karena spesies terlihat mirip.

- Peluang:

1) Data Iris bisa dijadikan dasar membuat sistem klasifikasi cepat dan akurat.

2) Bisa diterapkan untuk penelitian botani, perdagangan bunga, hingga supply chain.
